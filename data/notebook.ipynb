```python
from bs4 import BeautifulSoup

# Open the local HTML file and read it
with open("pandas/reference/series.html", "r", encoding="utf-8") as file:
    html_content = file.read()

# Parse the HTML using BeautifulSoup
soup = BeautifulSoup(html_content, "lxml")

# Print the prettified HTML (optional)
print(soup.prettify())
```

```python
import pandas as pd

col = soup.ul.text
answer = soup.section.text
df = pd.DataFrame({'Col': [col],'Answer': [answer]})
df.to_csv('cleaned/series.csv')
```
